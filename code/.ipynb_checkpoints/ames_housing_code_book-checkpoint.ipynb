# 5. Feature Engineering

### Dummy Variables

```python
# below is confirming the list of nominal features to make into dummies
features_dict['nominal']
```

```python
# dummy all nominal features
df_train_dummies = pd.get_dummies(df_train[features_dict['nominal']])
df_test_dummies = pd.get_dummies(df_test[features_dict['nominal']])
```

```python
# list of new test columns (have to exclude Sale Price)
df_dummies_test_new_cols = list(set(df_train_dummies.columns) - set(df_test_dummies.columns))

df_dummies_test_new_cols
```

```python
# list of new train cols
df_train_new_cols = list(set(df_test_dummies.columns) - set(df_train_dummies.columns))
df_train_new_cols
```

```python
# adding new dummy columns to test with value 0
for col in df_dummies_test_new_cols:
    df_test_dummies[col] = 0
```

```python
# adding new dummy columns to train with value 0
for col in df_train_new_cols:
    df_train_dummies[col] = 0
```

```python
list(set(df_train_dummies.columns) - set(df_test_dummies.columns))
```

We now have dummies list with all dummies for nominal features.  However, we now need to simulate the drop first method in the get dummies function.

```python
features_dict['nominal']
```

```python
df_train_dummies.shape
```

```python
'ms_subclass' in 'ms_subclass_2'
```

```python
dum_dict = {}
for feature in features_dict['nominal']:
    dum_dict[feature] = []
    for col in df_train_dummies.columns:
        if feature in col:
            dum_dict[feature].append(col)
```

```python
for key in dum_dict:
    df_train_dummies.drop(columns = dum_dict[key][0], inplace = True)
    df_test_dummies.drop(columns = dum_dict[key][0], inplace = True)
```

We can now merge our dummies columns with our data frames and drop the nominal features column.

```python
df_train = df_train.merge(df_train_dummies, on = 'Id')
df_test = df_test.merge(df_test_dummies, on = 'Id')
```

```python
df_train.drop(columns = features_dict['nominal'], inplace = True)
df_test.drop(columns = features_dict['nominal'], inplace = True)
```

### Interaction Features and Other Additions/Subtractions

First off, I want to create an interaction feature between the overall quality and then the condition.  I believe this may have a compounding effect.

```python
df_train['qual_cond_interaction'] = df_train['overall_qual'] * df_train['overall_cond']
df_test['qual_cond_interaction'] = df_test['overall_qual'] * df_test['overall_cond']
```

Second, instead of a feature called year built, I want to have a column with the age of the building.  We will follow the same for remodeling and garage. We will assume it is currently 2011 as our most recent home sale is 2010.  

```python
df_train['home_age'] = 2011 - df_train['year_built']
df_test['home_age'] = 2011 - df_test['year_built']
df_train['remod_age'] = 2011 - df_train['year_remod/add']
df_test['remod_age'] = 2011 - df_test['year_remod/add']
df_train['garage_age'] = 2011 - df_train['garage_yr_blt']
df_test['garage_age'] = 2011 - df_test['garage_yr_blt']
df_train.drop(columns = ['year_built', 'year_remod/add', 'garage_yr_blt'], inplace = True)
df_test.drop(columns = ['year_built', 'year_remod/add', 'garage_yr_blt'], inplace = True)
```

```python
df_train.head()
```

Similar to overall quality, we see features for basement quality and condition.  We should make an interaction variable for this. 

```python
df_train['bsmt_qual_cond_interaction'] = df_train['bsmt_qual'] * df_train['bsmt_cond']
df_test['bsmt_qual_cond_interaction'] = df_test['bsmt_qual'] * df_test['bsmt_cond']
df_train['garage_qual_cond_interaction'] = df_train['garage_qual'] * df_train['garage_cond']
df_test['garage_qual_cond_interaction'] = df_test['garage_qual'] * df_test['garage_cond']
```

# 7. Modeling

## Linear Regression

This is a simple linear regresssion taking all features as inputs to identify any interesting trends with the coefficients before moving forward with additional feature engineering.

```python
# setting features for stronger correlated features
features = [i for i in df_train if df_train[i].corr(df_train['saleprice']) > 0.3 or df_train[i].corr(df_train['saleprice']) < -0.3]
features.remove('saleprice')

# setting X and y variables
X = df_train[features]
y = np.log(df_train['saleprice'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 22)
```

```python
#instantiate model
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
cross_val_score(lr, X_train, y_train, cv=5).mean()
```

```python
lr_train_preds = np.exp(lr.predict(X_train))
lr_test_preds = np.exp(lr.predict(X_test))
```

```python
print(f'Model 1 Train RMSE: {np.sqrt(mean_squared_error(np.exp(y_train), lr_train_preds))}')
print(f'Model 1 Test RMSE: {np.sqrt(mean_squared_error(np.exp(y_test), lr_test_preds))}')
```

### Error Regression Plot

```python
resids = np.exp(y_test) - lr_test_preds
plt.scatter(np.exp(y_test), resids);
plt.axhline(0, color = 'red');
plt.xlabel('Testing Set Sale Prices');
plt.ylabel('Residuals');
plt.title('Linear Regression Residual Plot', size = 14);
```

## Lasso Regression

```python
# Set up a list of Lasso alphas to check.
l_alphas = np.logspace(-3, 0, 100)

# Cross-validate over our list of Lasso alphas.
lasso = LassoCV(alphas=l_alphas, cv=5, max_iter = 5000)

# Fit model using best ridge alpha!
lasso.fit(z_train, y_train)
```

## Ridge Regression

```python
# Set up a list of ridge alphas to check.
# Generates 100 values equally between 0 and 5,
# then converts them to alphas between 10^0 and 10^5.
r_alphas = np.logspace(0, 5, 100)

# Cross-validate over our list of ridge alphas.
ridge = RidgeCV(alphas = r_alphas, scoring = 'r2', cv = 5)

# Fit model using best ridge alpha!
ridge.fit(z_train, y_train)
```

```python
ridge.score(z_train, y_train)
```

```python
ridge.score(z_test, y_test)
```

```python
ridge_train_preds = np.exp(ridge.predict(z_train))
ridge_test_preds = np.exp(ridge.predict(z_test))

print(f'Model 1 Train RMSE: {np.sqrt(mean_squared_error(np.exp(y_train), ridge_train_preds))}')
print(f'Model 2 Test RMSE: {np.sqrt(mean_squared_error(np.exp(y_test), ridge_test_preds))}')
```

### Error Regression Plot

```python
resids = np.exp(y_test) - ridge_test_preds
plt.scatter(np.exp(y_test), resids);
plt.axhline(0, color = 'red');
plt.xlabel('Testing Set Sale Prices');
plt.ylabel('Residuals');
plt.title('Ridge Residual Plot', size = 14);
```
